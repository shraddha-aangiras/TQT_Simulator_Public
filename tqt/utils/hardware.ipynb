```python
import pyvisa
rm = pyvisa.ResourceManager()
print(rm.list_resources())
for resource in rm.list_resources():
    try:
        with rm.open_resource(resource) as instrument:
            print(f"Address = {resource}, instrument information: {instrument.query('*IDN?')}")
    except:
        pass
```
